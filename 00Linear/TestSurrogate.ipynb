# Linear System


## Discription


The studied Duffing system follows:

$$
\begin{equation}
    \ddot{x} = - p \hat{x} - d \hat{\dot{x}},
\end{equation}
$$

where $\hat{x}$ and $\hat{\dot{x}}$ refers to delayed(sampled) displacement and velocity.


## Import Functions and `.ode` Files

```python
import numpy as np
import pyxpp.pyxpp as pyxpp
import matplotlib.pyplot as plt
import sys
import os

sys.path.insert(0, os.path.abspath(".."))
from utils import *

plt.rcParams["font.serif"] = ["Times New Roman"]

Approx_Order1 = "ApproxSys/LinearApproxOrder1.ode"
Approx_Order3 = "ApproxSys/LinearApproxOrder3.ode"
Approx_Order20 = "ApproxSys/LinearApproxOrder20.ode"

Surrogate_Order1 = "SurrogateSys/SurrogateOrder1.ode"
Surrogate_Order3 = "SurrogateSys/SurrogateOrder3.ode"
Surrogate_Order20 = "SurrogateSys/SurrogateOrder20.ode"
```

## Simulation with `XPPaut`

Approximated systems

```python
SamplingTime = 0.31
SimulationTime = 20
Data_Approx_Order1 = pyxpp.run(Approx_Order1, ts=SamplingTime, total=SimulationTime)
Data_Approx_Order3 = pyxpp.run(Approx_Order3, ts=SamplingTime, total=SimulationTime)
Data_Approx_Order20 = pyxpp.run(Approx_Order20, ts=SamplingTime, total=SimulationTime)
```

Surrogate systems of the approximated systems

```python
Data_Surrogate_Order1 = pyxpp.run(
    Surrogate_Order1, epsilon=0.2, ts=SamplingTime, total=SimulationTime
)
```

## Visualization

```python
plt.plot(Data_Surrogate_Order1[:,1],Data_Surrogate_Order1[:,2])
plt.show()
```

```python
plt.plot(Data_Approx_Order1[:, 1], Data_Approx_Order1[:, 2], label="1st Order")
plt.plot(Data_Approx_Order3[:, 1], Data_Approx_Order3[:, 2], label="3rd Order")
plt.plot(Data_Approx_Order20[:, 1], Data_Approx_Order20[:, 2], label="20th Order")
plt.scatter(1, 0, c="black", marker="x", s=200, label="Initial Condition")
plt.xlabel(r"$x$", fontsize=12)
plt.ylabel(r"$\dot{x}$", fontsize=12, rotation=np.pi / 2)
plt.title("Phase Portrait", fontsize=15)
plt.legend()
plt.show()

plt.plot(Data_Approx_Order1[:, 0], Data_Approx_Order1[:, 1], label="1st Order")
plt.plot(Data_Approx_Order3[:, 0], Data_Approx_Order3[:, 1], label="3rd Order")
plt.plot(Data_Approx_Order20[:, 0], Data_Approx_Order20[:, 1], label="20th Order")
plt.xlabel(r"$t$", fontsize=12)
plt.ylabel(r"$x$", fontsize=12, rotation=np.pi / 2)
plt.title(r"$x-t$ Diagram", fontsize=15)
plt.legend()
plt.show()

plt.plot(Data_Approx_Order1[:, 0], Data_Approx_Order1[:, 2], label="1st Order")
plt.plot(Data_Approx_Order3[:, 0], Data_Approx_Order3[:, 2], label="3rd Order")
plt.plot(Data_Approx_Order20[:, 0], Data_Approx_Order20[:, 2], label="20th Order")
plt.xlabel(r"$t$", fontsize=12)
plt.ylabel(r"$\dot{x}$", fontsize=12, rotation=np.pi / 2)
plt.title(r"$\dot{x}-t$ Diagram", fontsize=15)
plt.legend()
plt.show()
```
