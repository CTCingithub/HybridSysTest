# Linear System


## Discription


The studied Duffing system follows:

$$
\begin{equation}
    \ddot{x} = - p \hat{x} - d \hat{\dot{x}},
\end{equation}
$$

where $\hat{x}$ and $\hat{\dot{x}}$ refers to delayed(sampled) displacement and velocity.


## Import Functions and `.ode` Files

```python
import numpy as np
import pyxpp.pyxpp as pyxpp
import matplotlib.pyplot as plt
import sys
import os

sys.path.insert(0, os.path.abspath(".."))
from utils import *

plt.rcParams["font.serif"] = ["Times New Roman"]
Approx_Order1 = "ApproxSys/LinearApproxOrder1.ode"
Surrogate_Order1 = "SurrogateSys/SurrogateOrder1.ode"
```

## Simulation with `XPPaut`

Approximated systems

```python
SamplingTime = 0.31
SimulationTime = 10
Data_ApproxOrder1 = pyxpp.run(Approx_Order1, ts=SamplingTime, total=SimulationTime)
Data_Surrogate_Order1_1 = pyxpp.run(
    Surrogate_Order1, epsilon=0.2, ts=SamplingTime, total=SimulationTime
)
Data_Surrogate_Order1_2 = pyxpp.run(
    Surrogate_Order1, epsilon=0.4, ts=SamplingTime, total=SimulationTime
)
Data_Surrogate_Order1_3 = pyxpp.run(
    Surrogate_Order1, epsilon=1, ts=SamplingTime, total=SimulationTime
)
Data_Surrogate_Order1_4 = pyxpp.run(
    Surrogate_Order1, epsilon=2, ts=SamplingTime, total=SimulationTime
)
Data_Surrogate_Order1_5 = pyxpp.run(
    Surrogate_Order1, epsilon=7.5, ts=SamplingTime, total=SimulationTime
)
```

Surrogate systems of the approximated systems

## Visualization

```python
plt.scatter(1, 0, c="black", marker="x", s=200, label="Initial Condition")
plt.plot(Data_ApproxOrder1[:, 1], Data_ApproxOrder1[:, 2], label="1st Order Expansion")
plt.plot(
    Data_Surrogate_Order1_1[:, 1],
    Data_Surrogate_Order1_1[:, 2],
    label=r"$\varepsilon = 0.2$",
)
plt.plot(
    Data_Surrogate_Order1_2[:, 1],
    Data_Surrogate_Order1_2[:, 2],
    label=r"$\varepsilon = 0.4$",
)
plt.plot(
    Data_Surrogate_Order1_3[:, 1],
    Data_Surrogate_Order1_3[:, 2],
    label=r"$\varepsilon = 1.0$",
)
plt.plot(
    Data_Surrogate_Order1_4[:, 1],
    Data_Surrogate_Order1_4[:, 2],
    label=r"$\varepsilon = 2.0$",
)
plt.plot(
    Data_Surrogate_Order1_5[:, 1],
    Data_Surrogate_Order1_5[:, 2],
    label=r"$\varepsilon = 7.5$",
)
plt.xlabel(r"$x$", fontsize=12)
plt.ylabel(r"$\dot{x}$", fontsize=12, rotation=np.pi / 2)
plt.title("Surrogate System of 1st Order Expansion", fontsize=15)
plt.legend()
plt.show()
```
