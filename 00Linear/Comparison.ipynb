# Comparison between 3 Systems


## 1. Discription


The studied Duffing system follows:

$$
\begin{equation}
    \ddot{x} = - p \hat{x} - d \hat{\dot{x}},
\end{equation}
$$

where $\hat{x}$ and $\hat{\dot{x}}$ refers to delayed(sampled) displacement and velocity.


## 2. Simulation with `XPPaut`


### 2.1 Import Functions and `.ode` Files


```python
import numpy as np
import pyxpp.pyxpp as pyxpp
import matplotlib.pyplot as plt
import sys
import os

sys.path.insert(0, os.path.abspath(".."))
from utils import *

plt.rcParams["font.serif"] = ["Times New Roman"]

Approx_Order1 = "ApproxSys/LinearApproxOrder1.ode"
Surrogate_Order1 = "SurrogateSys/SurrogateOrder1.ode"
ConstantDelaySys = "ConstantDelay.ode"
SamplingSys = "LinearSampling.ode"
```

### 2.2 Code Run


```python
SamplingTime = 0.31
SimulationTime = 20

Data_Sampling = pyxpp.run(SamplingSys, ts=SamplingTime, total=SimulationTime)
Data_ApproxOrder0 = pyxpp.run(
    ConstantDelaySys, ts=1.5 * SamplingTime, total=SimulationTime
)
Data_ApproxOrder1 = pyxpp.run(Approx_Order1, ts=SamplingTime, total=SimulationTime)
Data_SurrogateOrder1 = pyxpp.run(
    Surrogate_Order1, epsilon=1, ts=SamplingTime, total=SimulationTime
)
```

## 3. Visualization


```python
plt.plot(Data_Sampling[:, 1], Data_Sampling[:, 2], label="Sampling")
plt.plot(
    Data_ApproxOrder0[:, 1],
    Data_ApproxOrder0[:, 2],
    label="0th Order Expansion",
)
plt.plot(Data_ApproxOrder1[:, 1], Data_ApproxOrder1[:, 2], label="1st Order Expansion")
plt.title(
    "Phase Portrait when " + r"$t_{s} = $" + str(SamplingTime) + " s", fontsize=15
)
plt.legend()
plt.show()

plt.plot(Data_Sampling[:, 0], Data_Sampling[:, 2], label="Sampling")
plt.plot(
    Data_ApproxOrder0[:, 0],
    Data_ApproxOrder0[:, 2],
    label="0th Order Expansion",
)
plt.plot(Data_ApproxOrder1[:, 0], Data_ApproxOrder1[:, 2], label="1st Order Expansion")
plt.title(
    r"$x-t$" + " Diagram when " + r"$t_{s} = $" + str(SamplingTime) + " s", fontsize=15
)
plt.legend()
plt.show()
```

```python
plt.plot(
    Data_SurrogateOrder1[:, 1], Data_SurrogateOrder1[:, 2], label="1st Order Surrogate"
)
plt.title("Phase Portrait when " + r"$t_{s} = $" + str(SamplingTime) + " s")
plt.legend()
plt.show()
```
