# Linear System


## Discription


The studied Duffing system follows:

$$
\begin{equation}
    \ddot{x} = - p \hat{x} - d \hat{\dot{x}},
\end{equation}
$$

where $\hat{x}$ and $\hat{\dot{x}}$ refers to delayed(sampled) displacement and velocity.


## Import Functions and `.ode` Files

```python
import numpy as np
import pyxpp.pyxpp as pyxpp
import matplotlib.pyplot as plt
import sys
import os

sys.path.insert(0, os.path.abspath(".."))
from utils import *

plt.rcParams["font.serif"] = ["Times New Roman"]

ODE_Sampling = "LinearSampling.ode"
```

## Simulation with `XPPaut`

```python
SamplingTime = (0.25, 0.30, 0.35)
SimulationTime = 20
Data_Sampling_1 = pyxpp.run(
    ODE_Sampling, ts=SamplingTime[0], total=SimulationTime, dt=1e-5
)
Data_Sampling_2 = pyxpp.run(
    ODE_Sampling, ts=SamplingTime[1], total=SimulationTime, dt=1e-5
)
Data_Sampling_3 = pyxpp.run(
    ODE_Sampling, ts=SamplingTime[2], total=SimulationTime, dt=1e-5
)
```

## Visualization

```python
plt.plot(
    Data_Sampling_1[:, 1],
    Data_Sampling_1[:, 2],
    label=f"$t_s=${str(SamplingTime[0])}s",
)
plt.plot(
    Data_Sampling_2[:, 1],
    Data_Sampling_2[:, 2],
    label=f"$t_s=${str(SamplingTime[1])}s",
)
plt.plot(
    Data_Sampling_3[:, 1],
    Data_Sampling_3[:, 2],
    label=f"$t_s=${str(SamplingTime[2])}s",
)
plt.scatter(1, 0, c="black", marker="x", s=200, label="Initial Condition")
plt.xlabel(r"$x$", fontsize=12)
plt.ylabel(r"$\dot{x}$", fontsize=12, rotation=np.pi / 2)
plt.title("Phase Portrait", fontsize=15)
plt.legend()
plt.show()

plt.plot(
    Data_Sampling_1[:, 0],
    Data_Sampling_1[:, 1],
    label=f"$t_s=${str(SamplingTime[0])}s",
)
plt.plot(
    Data_Sampling_2[:, 0],
    Data_Sampling_2[:, 1],
    label=f"$t_s=${str(SamplingTime[1])}s",
)
plt.plot(
    Data_Sampling_3[:, 0],
    Data_Sampling_3[:, 1],
    label=f"$t_s=${str(SamplingTime[2])}s",
)
plt.xlabel(r"$t$", fontsize=12)
plt.ylabel(r"$x$", fontsize=12, rotation=np.pi / 2)
plt.title(r"$x-t$ Diagram", fontsize=15)
plt.legend()
plt.show()

plt.plot(
    Data_Sampling_1[:, 0],
    Data_Sampling_1[:, 2],
    label=f"$t_s=${str(SamplingTime[0])}s",
)
plt.plot(
    Data_Sampling_2[:, 0],
    Data_Sampling_2[:, 2],
    label=f"$t_s=${str(SamplingTime[1])}s",
)
plt.plot(
    Data_Sampling_3[:, 0],
    Data_Sampling_3[:, 2],
    label=f"$t_s=${str(SamplingTime[2])}s",
)
plt.xlabel(r"$t$", fontsize=12)
plt.ylabel(r"$\dot{x}$", fontsize=12, rotation=np.pi / 2)
plt.title(r"$\dot{x}-t$ Diagram", fontsize=15)
plt.legend()
plt.show()
```
